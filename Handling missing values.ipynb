```python
# libraries used

import pandas as pd
import numpy as np

# read the dataset file

sf_permits = pd.read_csv("Building_Permits.csv")

# set seed for reproducibility

np.random.seed(0)
```

```python
# looking at the dataset there are few rows missing data already
sf_permits.sample(5)
```

```python
# getting the number of missing data points per column
missing_values = sf_permits.isnull().sum()

# looking at number of missing points in the first 10 columns
missing_values[0:10]
```

That seems a lot of missing data. It will be helpful to see what percentage of values in our dataset are missing 

```python
total_cells = np.product(sf_permits.shape)
total_missing = missing_values.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

Wow more than quarter of the cells in the dataset are empty. Is this value msising because it wasn't recorded or because it doesn't exist ?

```python
missing_values[0:10]
```

```python
# remove all the rows that contain a missing value
sf_permits.dropna()
```

It looks like that's removed all our data. This is because every row in our dataset has a missing value. Now we will have to remove all the columns that have atleast one missing value

```python
# remove all columns with atleast one missing value
columns_with_na = sf_permits.dropna(axis=1)
columns_with_na.head()
```

```python
# comparing the two dataset to check how much data we lost 
print("Columns in original dataset: %d \n" % sf_permits.shape[1])
print("Columns with na's dropped: %d" % columns_with_na.shape[1])
```

```python
subset = sf_permits.loc[:, 'Permit Number':'Record ID'].head()
subset
```

```python
# replacing na's with 0
subset.fillna(0)
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the reamining na's with 0
sf_permits.fillna(method = 'bfill', axis=0).fillna(0)
```

Fillig missing value is also know as imputation. So either you can fill it with "0" or else you can use the values that are below and above of the missing values.
